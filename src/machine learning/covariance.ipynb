# Covariance

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dataset
N = 1000
np.random.seed(42069)
x = np.random.rand(N)
y = np.random.rand(N) + 5 * x
colors = np.random.rand(N)
```

```python
data = np.array([x,  y])
matrix = np.cov(data, bias = True)
print(matrix)
```

```python
# plot x and y with color as a scatter plot
plt.scatter(x, y, c = colors, alpha = 0.5)
```
