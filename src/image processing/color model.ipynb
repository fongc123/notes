# Image Processing: Color Model

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in `image.png`
img = cv2.imread('image.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# get RGB channels
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
```

```python
# plot individual RGB channels
f, axes = plt.subplots(1, 3, figsize=(10, 4))
f.suptitle("RGB Channels", fontsize = 16)
for i in range(3):
    axes[i].set_title(['Red', 'Green', 'Blue'][i])
    axes[i].imshow(img[:, :, i], cmap = [ "Reds", "Greens", "Blues"][i])
    axes[i].axis('off')
f.tight_layout()
plt.show()
```

```python
# plot HSI channels
hsi = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
f, axes = plt.subplots(1, 3, figsize = (10, 4))
f.suptitle("HSI Channels", fontsize = 16)
for i in range(3):
    axes[i].set_title(['Hue', 'Saturation', 'Intensity'][i])
    axes[i].imshow(hsi[:, :, i], cmap = [ "hsv", "gray", "gray"][i])
    axes[i].axis('off')
f.tight_layout()
plt.show()
```
