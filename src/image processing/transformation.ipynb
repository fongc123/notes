# Image Processing: Intensity Transformations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in image.png and convert to grayscale
FILENAME = "image.png"
img = cv2.imread(FILENAME, cv2.IMREAD_GRAYSCALE)

# create window
cv2.namedWindow('window', cv2.WINDOW_NORMAL)
cv2.resizeWindow("window", 600, 600)

# show image
cv2.imshow('window', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def show_histogram(img):
    # change figure size
    plt.figure(figsize=(12, 5))

    # show original image
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')

    # plot image histogram
    plt.subplot(1, 2, 2)
    plt.hist(img.ravel(), 256, [0, 256])
    plt.xlim([0, 256])
    plt.title("Histogram of original image")
    plt.xlabel("Intensity")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
print("Image shape:", img.shape)
```

```python
show_histogram(img)
```

```python
# apply histogram equalization
img_eq = cv2.equalizeHist(img)
show_histogram(img_eq)
```

```python
# apply power-law transformation with gamma > 1
gamma_level = 1.5
img_gamma_pos = np.power(img / 255, gamma_level) * 255
img_gamma_pos = img_gamma_pos.astype(np.uint8)
show_histogram(img_gamma_pos)
```

```python
# apply power-law transformation with gamma < 1
gamma_level = 0.5
img_gamma_neg = np.power(img / 255, gamma_level) * 255
img_gamma_neg = img_gamma_neg.astype(np.uint8)
show_histogram(img_gamma_neg)
```

```python
# apply log transformation
img_log = np.log(img + 1) * 255 / np.log(256)
img_log = img_log.astype(np.uint8)
show_histogram(img_log)
```

```python
# apply negative transformation
img_neg = 255 - img
show_histogram(img_neg)
```
