# Image Processing: Fourier Transform

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
def plot_input_spectrum(original, spectrum):
    # plots the image and its spectrum
    plt.subplot(121), plt.imshow(original, cmap = 'gray')
    plt.title('Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(spectrum, cmap = 'gray')
    plt.title("Spectrum"), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
# load image and convert to grayscale
FILENAME = "image.png"
img = cv2.imread(FILENAME, cv2.IMREAD_GRAYSCALE)
```

```python
img.shape
```

```python
# apply fourier transform
f = np.fft.fft2(img)
original_shift = np.fft.fftshift(f)
original_spectrum = np.log(1 + np.abs(original_shift))

# plot magnitude spectrum
plot_input_spectrum(img, original_spectrum)
```

```python
# apply inverse fourier transform
ishift = np.fft.ifftshift(original_shift)
if_shift = np.fft.ifft2(ishift)
img_back = np.abs(if_shift)

# plot inverse fourier transform
plot_input_spectrum(img_back, original_spectrum)
```
