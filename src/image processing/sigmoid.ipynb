# Sigmoid Function Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot sigmoid function
with plt.xkcd(randomness = 2):
    alpha1 = 0.1
    beta1 = 100
    plt.subplot(1, 2, 1)
    x = np.arange(0, 255, 0.1)
    y = 1 / (1 + np.exp(-alpha1 * (x - beta1)))
    plt.plot(x, y)
    plt.xlabel('Input intensity')
    plt.ylabel('Sigmoid output')
    plt.title('$\\alpha$={}, $\\beta$={}'.format(alpha1, beta1))

    alpha2 = 20
    beta2 = 100
    plt.subplot(1, 2, 2)
    x = np.arange(0, 255, 0.1)
    y = 1 / (1 + np.exp(-alpha2 * (x - beta2)))
    plt.plot(x, y)
    plt.xlabel('Input intensity')
    plt.ylabel('Sigmoid output')
    plt.title('$\\alpha$={}, $\\beta$={}'.format(alpha2, beta2))
    plt.tight_layout()

    plt.show()
```
